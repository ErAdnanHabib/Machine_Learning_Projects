```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
data_set = pd.read_csv('insurance.csv')
```

```python
data_set.info()
data_set
```

```python
data_set.region.value_counts()
```

```python
#NOW AS TEXT MATERIAL CANNOT BE LEARNED BY ML MODEL WE HAVE TO TRANSFORM THE TEXT PART I.E SEX , SMOKER , REGION IN MATHEMATICAL FUNCTIONS

data_set.replace({'sex' : {'male':0 , 'female':1}},inplace=True)
data_set.replace({'smoker' : {'yes':0 , 'no':1}},inplace=True)
data_set.replace({'region' : {'southeast':0 , 'southwest':1 , 'northwest':2 , 'northeast':3}}, inplace=True)
data_set
```

```python
x=data_set.drop(columns = 'charges' , axis=1) #data
y=data_set['charges']  #target
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.20 , random_state=0)
```

```python
st_x = StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

```python
clf = LinearRegression()
clf.fit(x_train , y_train)
```

```python
y_pred = clf.predict(x_test)
r2_pred=metrics.r2_score(y_test , y_pred)*100
```
